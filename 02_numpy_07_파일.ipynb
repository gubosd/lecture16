### genfromtxt()
- loadtxt() 보다 세밀한 기능 제공

````python
help(np.genfromtxt)
````

````python
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

iris = np.genfromtxt('iris.csv', skip_header=1, delimiter=',',
                     converters={4: lambda s: float(labels.index(s.decode()))})
iris.shape
````

### savetxt()

````python
help(np.savetxt)
````

````python
iris = np.loadtxt('iris.csv', skiprows=1, delimiter=',', usecols=[0,1,2,3])
````

````python
np.savetxt('iris2.csv', iris, delimiter='파이썬') # 윈도우에서는 인코딩이 cp949 로 적용됨 
````

````python
np.savetxt('iris2.csv', iris, delimiter=',', fmt='%.2f', encoding='utf-8') # 윈도우에서는 인코딩이 cp949 로 적용됨 
````
